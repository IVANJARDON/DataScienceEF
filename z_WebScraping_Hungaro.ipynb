```python
import pandas as pd
import numpy as np
import requests
import bs4
```

```python
url = 'http://www.hungarianalgorithm.com/solve.php?c=4400-11800-16200-37000-53000-70000-90100--0-8200-12400-32200-47800-64000-83500--1000000-0-4800-22600-37400-52000-70300--1000000-1000000-0-17800-32200-46000-63700--1000000-1000000-1000000-0-14000-25000-40600--1000000-1000000-1000000-1000000-0-10000-24100--1000000-1000000-1000000-1000000-1000000-0-14200'
res = requests.get(url)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
a = soup.find_all('table')[1].find_all('tr')[0].find_all('td')[0].text.replace(',','')
int(a)
```

```python
n = 7
total = pd.DataFrame()
for table in soup.find_all('table')[1:]:
    values = []
    for row in table.find_all('tr'):
        for col in row.find_all('td'):
            try:
                values.append(int(col.text.replace(',','')))
            except:
                continue
    tabla = pd.DataFrame(np.reshape(np.array(values[:n**2]),(n,n)))
    display(tabla)
    total = total.append(['.']).append(tabla).fillna('.').replace('.','')

total.to_csv('Tablas_hungaro.csv')
```
