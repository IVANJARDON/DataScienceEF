## PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
pca2 = PCA(n_components = 2)
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), 
                  index = df.index)
Xp2 = pd.DataFrame(pca2.fit_transform(df),
                   index = df.index, 
                   columns = ['p1','p2'])
Xp2.reset_index(inplace = True)
Xp2['label'] = Xp2['label'].astype(str)
sns.lmplot(data = Xp2.sample(frac = 0.05),
           x = 'p1',
           y = 'p2',
           fit_reg = False, 
           hue = 'label')
```

```python
pca = PCA(.9)
X = pd.DataFrame(pca.fit_transform(df), index = df.index)
y = X.index.values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = 0.77, 
                                                    random_state = 22)
```

## Modelo

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

param_logreg = {'penalty':['l1', 'l2', 'elasticnet'], 
                'C':[x for x in range(1,5)], 
                'class_weight':['None','balanced'],
                'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
                }

from sklearn.model_selection import RandomizedSearchCV
modelo = RandomizedSearchCV(param_distributions = param_logreg, 
                            cv = 3, 
                            n_jobs = -1, 
                            scoring = 'accuracy',
                            estimator = logreg,
                            verbose = 5,
                            n_iter = 5,
                            random_state = 22)

modelo.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm = pd.DataFrame(confusion_matrix(y_true = y_test,
                                   y_pred = modelo.predict(X_test)))

## Tiene buena acertividad (suma de diagonal en la matriz de confusión)
display('Accuracy de {:.2%}'.format(np.asarray(cm).trace()/len(y_test)))

fig, ax = plt.subplots(figsize=(7,7)) 
sns.heatmap(pd.DataFrame([cm[col]/cm[col].sum() for col in cm.columns]), 
            annot = True,
            fmt = '.0%',
            cmap = 'Blues',
            linewidths = 0.5, 
            ax = ax)
```
